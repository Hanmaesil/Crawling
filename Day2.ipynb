# 한달동안의 영화 데이터 수집

```python
1월 1일
20220101 ~ 20220106
```

```python
title_list = []
score_list = []
rank_list = []

for i in range(20220101,20220107) :
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=' + str(i)
    res = req.get(url)
    soup = bs(res.text,'lxml')
    title = soup.select('div.tit5 > a')
    score = soup.select('td.point')
    
 

    for i in range(len(title)) :
        title_list.append(title[i].text)
        score_list.append(score[i].text)
        rank_list.append(i+1)
```

```python
len(title_list)
```

```python
dic = {'영화제목' : title_list, "평점" : score_list, "순위" : rank_list}
movie = pd.DataFrame(dic)
movie.set_index('순위', inplace = True)
movie
```

# Pandas 활용하여 날짜 생성

```python
# pd.date_range(start = '시작하는 날짜', end = '끝나는 날짜')
date = pd.date_range(start = '2021-01-01', end = '2021-03-31') #적는 형태를 지켜야한다. 
```

```python
#문자형 포맷팅을 사용하여 '-' 없애보기
days = date.strftime('%Y%m%d')  # strftime -> string formating time 의 줄임말
#Y가 소문자일때
date.strftime('%y%m%d')
```

```python
days
```

```python
#진행속도를 보여준다
from tqdm import tqdm_notebook
tqdm_notebook()
```

```python
title_list = []
score_list = []
rank_list = []
day_list = []

for day in tqdm_notebook(days) :
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=' + day
    res = req.get(url)
    soup = bs(res.text,'lxml')
    title = soup.select('div.tit5 > a')
    score = soup.select('td.point')
    
 

    for i in range(len(title)) :
        title_list.append(title[i].text)
        score_list.append(score[i].text)
        rank_list.append(i+1)
        day_list.append(day)
```

```python
dic = {'영화제목' : title_list, "평점" : score_list, "순위" : rank_list, "날짜" : day_list}
movie = pd.DataFrame(dic)
movie.set_index('순위', inplace = True)
movie
```
