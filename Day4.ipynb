```python
#ë°˜ë³µë¬¸ì„ í†µí•œ ì‘ì—…!
item_list = []
price_list = []
driver = wb.Chrome("C:/Users/smhrd/í¬ë¡¤ë§/chromedriver.exe", options = options) # ì²˜ìŒ ì˜µì…˜ì€ ì†ì„±ì´ê³  =ë’¤ì˜ ì˜µì…˜ì€ ë³€ìˆ˜ëª…ì´ë‹¤
time.sleep(0.5)
driver.get("http://corners.gmarket.co.kr/Bestsellers")
time.sleep(0.5)
for i in range(3) :
    img = driver.find_elements_by_css_selector("img.lazy")
    img[i].click()
    time.sleep(1)
    #item = driver.find_element_by_css_selector(".itemtit")
    item_list.append(driver.find_element_by_css_selector(".itemtit").text)
    price = driver.find_element_by_css_selector(".price_real")
    price_list.append(price.text)
    driver.back()
    time.sleep(1)
dic = {"ìƒí’ˆëª…" : item_list, "ê°€ê²©" : price_list}
fin = pd.DataFrame(dic)
fin
fin.to_csv("gë§ˆì¼“.csv", encoding="euc-kr")
#ë°ì´í„° í”„ë ˆì„ ì œì‘
#csví˜•ì‹ìœ¼ë¡œ ì €ì¥
```
