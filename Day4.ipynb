# headless
- 크롬창을 실행시키지 않고 백그라운드에서 코드가 작동하게 해준다
- 기능을 사용하려면 드라이버 경로를 명시해 주어야 한다!!
- 폴더경로는 영어명으로 하자!!!!!!! -> 예시에서는 smhrd로 잘 되어있다

```python
options = wb.ChromeOptions()
options.add_argument('headless')
options.add_argument('window-size=1920x1080')
options.add_argument("disable-gpu")
```

```python
#반복문을 통한 작업!
item_list = []
price_list = []
driver = wb.Chrome("C:/Users/smhrd/크롤링/chromedriver.exe", options = options) # 처음 옵션은 속성이고 =뒤의 옵션은 변수명이다
time.sleep(0.5)
driver.get("http://corners.gmarket.co.kr/Bestsellers")
time.sleep(0.5)
for i in range(3) :
    img = driver.find_elements_by_css_selector("img.lazy")
    img[i].click()
    time.sleep(1)
    #item = driver.find_element_by_css_selector(".itemtit")
    item_list.append(driver.find_element_by_css_selector(".itemtit").text)
    price = driver.find_element_by_css_selector(".price_real")
    price_list.append(price.text)
    driver.back()
    time.sleep(1)
dic = {"상품명" : item_list, "가격" : price_list}
fin = pd.DataFrame(dic)
fin
fin.to_csv("g마켓.csv", encoding="euc-kr")
#데이터 프레임 제작
#csv형식으로 저장
```
